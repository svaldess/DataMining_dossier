```
import random
data = {}
numberOfBuckets = 10

f = open("./data.txt")
lines = f.readlines()
f.close()
# process trainfiles
for line in lines[1:]:
  line = line.replace(',', '\t')
 # fields = line.split()
  category = line.split()[12]
  data.setdefault(category, [])
  data[category].append(line)
#print(data)

buckets = []
for i in range(1,11):
  buckets.append([])
data['0'][0]
```

```
for k in data.keys():
  #randomize order of instances for each class
  random.shuffle(data[k])
  idx = 0
  # divide into buckets
  for item in data[k]:
    buckets[idx].append(item)
    idx = (idx + 1) % numberOfBuckets 
buckets[1]
```

```
for idx in range(numberOfBuckets):
  f = open("%s-%02i" % ("./data/dataset", idx+1), 'w')
  for item in buckets[idx]:
    f.write(item)
  f.close()
```
